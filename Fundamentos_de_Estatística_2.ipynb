## **2. Variáveis Qualitativas - Qui-quadrado**

Professor abordou apenas o teste qui-quadrado nesta aula para este tema. Ele fez de forma manual no excel, eu transformei em python de três formas.

1.   Modo 1: De forma extensa como se estivéssemos fazendo com lapis e papel para enteder de onde vem os resultados;

O objetivo é avaliar se existe relação entre as variáveis qualitativas ou categóricas. Conforme Fávero e Belfiore (2021), uma tabela de contingência também é chamada de tabela de classificação cruzada (*cross-tabulation*) ou ainda tabela de correspondência.

Resumo: O teste qui-quadrado (χ2) de Pearson (ou teste chi-quadrado de Pearson) é um teste estatístico aplicado a dados categóricos para avaliar quão provável é que qualquer diferença observada aconteça ao acaso.

**Exemplo 1 - Fávero e Belfiore (2021):** Um estudo foi realizado com 200 pessoas com o intuito de analisar o comportamento conjunto da variável “operadora de plano de saúde” (X) com a variável “nível de satisfação” (Y) do consumidor. O objetivo é analisar se existe a associação estatisticamente significativa entre tais variáveis.

*Modo 1: Da forma extensa*

```python
# Abertura de bibliotecas
import numpy as np
import pandas as pd
import scipy.stats

# Abertura e organização dos dados
dataset1 = pd.read_stata('/content/drive/MyDrive/Códigos Python MBA/Fundamentos de Estatística 2/Dados1.dta')

# Aparecendo as cinco primeiras posições do Dataset (ou banco de dados)
dataset1.head()
```

```python
# Passo 1: Calcule as Frequências absolutas observadas - tabela de contingência
freq_abs_obs = pd.crosstab(dataset1['operadora'],
                           dataset1['satisfacao'],
                           margins=True,
                           margins_name='Total')
freq_abs_obs
```

Para realizar os cálculos, os dados serão trabalhados em forma de matrizes utilizando o Pacote Numpy.

```python
# Passo 2: Calcule as frequências absolutas esperadas
b11 = (68*96)/200
b12 = (68*72)/200
b13 = (68*32)/200

b21 = (72*96)/200
b22 = (72*72)/200
b23 = (72*32)/200

b31 = (60*96)/200
b32 = (60*72)/200
b33 = (60*32)/200

# Frequências absolutas observadas
freq_abs_obs.to_numpy()

# Frequencias absolutas esperadas
freq_abs_esp = np.array([[b11, b12, b13],
                         [b21, b22, b23],
                         [b31, b32, b33]])
freq_abs_esp
```

```python
# Passo 3: Calcule a estatística teste qui-quadrado.
# chi2_individual = (Valor Observado - Valor Esperado)²/Valor esperado
# Arredondei para 2 casas decimais usando o comando 'round" e use sempre quando necessário.

q11 = round((40-b11)**2/b11,2)
q12 = round((16-b12)**2/b12,2)
q13 = round((12-b13)**2/b13,2)

q21 = round((32-b21)**2/b21,2)
q22 = round((24-b22)**2/b22,2)
q23 = round((16-b23)**2/b23,2)

q31 = round((24-b31)**2/b31,2)
q32 = round((32-b32)**2/b32,2)
q33 = round(( 4-b33)**2/b33,2)

# Qui-quadrado
qui2 = np.array([[q11, q12, q13],
                [q21, q22, q23],
                [q31, q32, q33]])
qui2
```

```python
# Passo 4: Somando todos os qui-quadrado individuais
qui2_calc = q11 + q12 + q13 + q21 + q22 + q23 + q31 + q32 + q33

# Imprima valor na tela
print("O valor de qui-quadrado é: ", qui2_calc)
```


**Aplicação do teste qui-quadrado para duas amostras independentes**

*   H0: Variáveis não são associadas
*   H1: Variáveis são associadas significativamente

Atenção: O nível de significância a ser considerado é de 5%

Consulte tabela da distribuição qui-quadrado em: https://www.ime.unicamp.br/~cnaber/Tabela%20da%20Qui-quadrado.pdf

```python
# Passo 5: Cálculo dos graus de liberdade sendo (i) linhas e (j) colunas, tem-se: gl = (i-1)*(j-1)
gl = (3-1)*(3-1)
print(gl)
```

```python
# Passo 6: Encontre o valor de qui-quadrado teórico.
# Faça através da interseção de gl = 4 e coluna "0.05" que significa 5% de significância na consulta da tabela
# Coloque em 'qui2_teórico' o valor observado na tabela
qui2_teorico = 9.488
print('Na tabela o qui2_teorico ou valor crítico é: ', qui2_teorico )


# Confirmação do qui-teórico via código.
# Obtendo valor crítico do qui-quadrado
print('Por código, o valor do qui2_teórico ou valor crítico é: ', round(scipy.stats.chi2.ppf(1-0.05, df = 4), 3))


# Confirmação de que o valor mostrado é realmente de 5% de significância (ou 95% de confiança)
# Obtendo nível de significancia do teste.
print('O nível de confiança deste teste é de: ', round(scipy.stats.chi2.cdf(9.488, df = 4), 3))
```

```python
# Passo 7: Teste Lógico de decisão pelos valores de qui-quadrado
if qui2_teorico > qui2_calc:
  print("Não rejeita H0")
else:
  print("Rejeita H0")
```

**Conclusão:** Ao nível de significância de 5%, rejeita-se H0.
Portanto, as variáveis são associadas.

**Poderia ter feito isso mais rápido? Sim, utilizando o Pacote Scipy como segue abaixo:**

*Modo 2: Modo Scipy*

```python
# Importando bibliotecas
import pandas as pd
import scipy.stats as stats

# Abertura e organização dos dados
dataset1 = pd.read_stata('/content/drive/MyDrive/Códigos Python MBA/Fundamentos de Estatística 2/Dados1.dta')

# Frequências absolutas observadas - tabela de contingência
freq_abs_obs = pd.crosstab(dataset1['operadora'],
                           dataset1['satisfacao'])

# Teste Qui-quadrado
chi2, p, dof, expected= stats.chi2_contingency(freq_abs_obs)

# Imprima na tela
print('Estatística qui-quadrado é: ', round(chi2, 2))
print('O p-valor é: ', round(p,3))
print('Graus de liberdade é: ', dof)
print('Frequências absolutas esperadas: \n', expected)
```

```python
# Teste Lógico de decisão pelo p-valor
if p > 0.05:
  print("Não rejeita H0")
else:
  print("Rejeita H0")
```

**Conclusão:** Ao nível de confiança de 95%, as variáveis são associadas em pelo menos um par de categorias.

Em seguida, deseja calcular os pares de correlações entre as notas de matemática – física, matemática – literatura e física – literatura. Quais são as correlações de Pearson obtidas?

```python
# Abertura de bibliotecas
import numpy as np
import pandas as pd
import scipy.stats

# Abertura e organização dos dados
dataset2 = pd.read_csv('/content/drive/MyDrive/Códigos Python MBA/Fundamentos de Estatística 2/Dados2.txt', sep='\t')

# Aparecendo as cinco primeiras posições do Dataset (ou banco de dados)
dataset2.head()
```

**3.1 Covariância**

Vamos calcular a covariânca de um par de variáveis do dataset de duas formas:

*   1.1   Usando a fórmula
*   1.2   Usando o comando

formula covariancia.jpeg

```python
# Passo 1: Vamos utilizar as médias para cada disciplina, respectivamente:
media_mat = dataset2['Nota Matemática'].mean()
media_fis = dataset2['Nota Física'].mean()
media_lit = dataset2['Nota Literatura'].mean()

# Imprima na tela
print('As médias de matemática, física e literatura são, respectivamente: \n',
      round(media_mat,2), '\n',
      round(media_fis,2), '\n',
      round(media_lit,2)
      )
```

```python
# Passo 2: Exemplo de como calcular a covariância a partir da fórmula entre "Nota Matemática" e "Nota Física"

dataset2['desvio_mat'] = dataset2['Nota Matemática']- media_mat         # (Xi - Xmedia)
dataset2['desvio_fis'] = dataset2['Nota Física']- media_fis             # (Yi - Ymedia)
dataset2['mat x fis'] = dataset2['desvio_mat'] * dataset2['desvio_fis'] # (Xi - Xmedia)*(Yi - Ymedia)
cov_mat_fis = round(dataset2["mat x fis"].sum()/(30-1),6)               # Soma a multiplicação dos desvios e divide por n-1 (Aqui: n=30)

# Mostre na tela a tabela de cálculo e o valor da covariância entre "Nota Matemática" e "Nota Física"
display(dataset2.head())
print('O valor da covariância entre as notas de matemática e física é: ', cov_mat_fis)
```

```python
# Passo 3: Vamos calcular a covariância de todos os pares de mat-fis, mat-lit, fis-lit via comando

# Removendo as colunas do cálculo de exemplo
dataset2 = dataset2.drop(dataset2.columns[[3,4,5]], axis=1)

# Calculando a covariância para todos os pares de variáveis do dataset2
covariancia = dataset2.cov()

# Imprima na tela
display(covariancia)
```

**3.2 Coeficiente de Correlação de Pearson**

*   1.2 Usando o comando

## **4. Distribuições de Probabilidades - Variável Discreta**

A caracteristica primordial antes da identificação da distribuição de probabilidade mais ajustada para variável que se tem no banco de dados, é identificar o tipo de variável que se está trabalhando.

*   **Variável aleatória discreta:** composta de valores inteiros;
    As distribuições de probabilidades permitidas para esse tipo de  variável são:



*   Uniforme
*   Bernoulli
*   Binomial
*   Binomial negativa
*   Poisson

*Distribuição Uniforme:* todos os possíveis valores tem a mesma probabilidade de ocorrência. Com valor de n representando a quantidade de possíveis valores. Equação:

uniforme.png

*Distribuição de Bernoulli*: Os valores da variável podem assumir apenas dois resultados possíveis chamados de sucesso (p) ou fracasso (1-p) quando ocorre apenas um experimento. É usado na logística binária. Equação:

bernoulli.png


*Distribuição Binomial*: ocorre quando há n repetições independentes do experimento de Bernoulli e a probabilidade de sucesso (p) é constante em todas as repetições. É usado na logística multinomial. A variável no modelo binomial indica a quantidade de sucessos (k) nas (n) repetições do experimento. Calcula-se a probabilidade de k sucessos dada a quantidade de ensaios realizados.

binomial.png

Exemplo: Em uma indústria, sabe-se que a probabilidade (p) de encontrar
peças defeituosas em cada lote produzido é 6,50%. São produzidos 12 lotes
(n) da peça por mês. Analise as seguintes probabilidades (k):

*   a) Qual a probabilidade de encontrar peças defeituosas em 2 lotes no mês?
*   b) Qual a probabilidade de encontrar peças defeituosas em 4 lotes no mês?
*   c) Qual a probabilidade de encontrar peças defeituosas em no máximo 2 lotes?

```python
# Parâmetros da distribuição binomial
# p (probabilidade de sucesso) = 6,50%
# n (número de repetições) = 12
# k (quantidade de sucessos de interesse)
```

```python
# Dentro do módulo math existe uma função factorial para calcular o fatorial de um número.
from math import factorial

# Parâmetros do exercicio da letra a
n = 12
k = 2
p = 0.065

# Usando a fórmula completa da distribuição binomial
# Multiplicação por 100 para transformar em % e arrendondamento para 2 casas
letra_A = round((factorial(n) / (factorial(k) * factorial(n-k)) * ((p)**k)*(1-p)**(n-k))*100, 2)

# Imprima resultado na tela
print('P(k=2) é: ', letra_A)
```

```python
# Dentro do módulo math existe uma função factorial para calcular o fatorial de um número.
from math import factorial

# Parâmetros do exercicio da letra b
n = 12
k = 4
p = 0.065

# Usando a fórmula completa da distribuição binomial
# Multiplicação por 100 para transformar em % e arrendondamento para 2 casas
letra_B = round((factorial(n) / (factorial(k) * factorial(n-k)) * ((p)**k)*(1-p)**(n-k))*100, 2)

# Imprima resultado na tela
print('P(k=4) é: ', letra_B)
```

```python
# Dentro do módulo math existe uma função factorial para calcular o fatorial de um número.
from math import factorial

# Parâmetros do exercicio da letra c
n = 12
p = 0.065
k1 = 0
k2 = 1
k3 = 2

# Usando a fórmula completa da distribuição binomial;
# Multiplicação por 100 para transformar em % e arrendondamento para 2 casas
C1 = round(((factorial(n) / (factorial(k1) * factorial(n-k1))) * ((p)**k1)*(1-p)**(n-k1))*100, 2)
C2 = round(((factorial(n) / (factorial(k2) * factorial(n-k2))) * ((p)**k2)*(1-p)**(n-k2))*100, 2)
C3 = round(((factorial(n) / (factorial(k3) * factorial(n-k3))) * ((p)**k3)*(1-p)**(n-k3))*100, 2)

# Somando as probabilidades
letra_C = C1 + C2 + C3

# Imprima resultado na tela
print('P(k=0) + P(k=1) + P(k=2) é: ', round(letra_C,2))
```

*Distribuição Binomial negativa*: Na distribuição binomial negativa, são realizados (x) ensaios independentes de Bernoulli até que sejam obtidos (k) sucessos. A probabilidade de sucesso (p) é constante em todos os ensaios realizados. A variável do modelo binomial negativa indica a quantidade de ensaios (x).
Ou seja: Quantos ensaios (x) precisam ser realizados para que se obtenha k sucessos?

Binomial negativa.png

Exemplo: Em um parque de diversões, existe uma máquina em que o jogador
deve capturar algum item utilizando os comandos de um braço mecânico.
Considere que a probabilidade (p) de que o jogador consiga capturar algum
item em cada jogada é 11%. Identifique as seguintes probabilidades:

*   a) De que o jogador necessite de 10 jogadas para capturar 3 itens.
*   b) De que o jogador necessite de 20 jogadas para capturar 3 itens.
*   c) De que o jogador necessite de 5 jogadas para capturar 1 item.

```python
# Parâmetros da distribuição binomial negativa
# p (probabilidade de sucesso) = 11%
# x (número de ensaios) = 12
# k (quantidade de sucessos de interesse)
```

```python
# Dentro do módulo math existe uma função factorial para calcular o fatorial de um número.
from math import factorial

# Parâmetros do exercicio da letra a
p = 0.11
x = 10
k = 3

# Usando a fórmula completa da distribuição binomial negativa
# Multiplicação por 100 para transformar em % e arrendondamento para 2 casas
A = round(((factorial(x-1)/(factorial(k-1)*factorial((x-1)-(k-1)))) * (p**k) *(1-p)**(x-k))*100,2)

# Imprima resultado na tela
print('P(X=10) e K=3 é: ', A)
```

Exemplo: Um médico notou que a taxa média de ocorrência (𝝀) de pacientes
com certa doença rara em seu consultório é de 2 por ano. Aceitando que esta
variável tenha distribuição Poisson, estime:

```python
# Dentro do módulo math existe uma função factorial e o número de Euler ou e para usar na distribuição de Poisson.
from math import factorial, e

# Parâmetros do exercicio da letra a
lamb_da = 2
k = 1

# Usando a fórmula completa da distribuição de Poisson
# Multiplicação por 100 para transformar em % e arrendondamento para 2 casas

A = round(((e**(-lamb_da) * lamb_da**k) / factorial(k))*100,2)

# Imprima resultado na tela
print('P(X=1) é: ', A)
```

## **5. Distribuições de Probabilidades - Variável Contínua**

*   Variável aleatória contínua: qualquer valor contido nos números reais;
    As distribuições de probabilidades permitidas para esse tipo de  variável são:
       *   Normal (Normal Padrão)
       *   Qui-quadrado
       *   t de Student
       *   F de Snedecor

*Distribuição Normal:* É a distribuição Gaussiana, com curva em forma de sino. Os parâmetros relevantes da distribuição normal são a média (μ) e o desvio padrão (σ) da variável

*Distribuição normal padrão:* para obter a normal padrão, transforma-se a variável por meio do Z-score. A transformação pelo Z-score faz com que a variável passe a ter média = 0 e desvio padrão = 1 e não altera a distribuição original.

*Distribuição qui-quadrado:* A distribuição χ2 apresenta curva assimétrica e positiva. É formada a partir da soma dos quadrados de v variáveis aleatórias independentes com distribuição normal padrão.

*Distribuição t de Student:* é parecida com a distribuição normal padrão, ou seja, tem formato de sino e é simétrica em torno da média. Por outro lado, a distribuição t de Student tem caudas mais longas e isso permite valores mais extremos. Uma aplicação da distribuição t de Student é o teste de hipóteses para médias

*Distribuição F de Snedecor:* Uma aplicação comum da distribuição  F(distribuição de Fischer) é a análise de variâncias. Tem distribuição de probabilidades assimétrica e positiva quando os graus de liberdade v1 e v2 são pequenos
